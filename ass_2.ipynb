```python
# prompt: Find the mean of a numeric column grouped by a categorical column in a 2D numpy array

import numpy as np
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
species = np.genfromtxt(url, delimiter=',', dtype='str', usecols=[4])

# Assuming you want the mean of the sepal length (column 0) grouped by species
unique_species = np.unique(species)
means = []
for s in unique_species:
  indices = np.where(species == s)
  sepal_lengths = iris_2d[indices, 0]
  mean_sepal_length = np.mean(sepal_lengths)
  means.append(mean_sepal_length)

print(f"Species: {unique_species}")
print(f"Mean Sepal Lengths: {means}")
```
